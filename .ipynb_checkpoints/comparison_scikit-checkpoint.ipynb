```python
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
df=pd.read_csv("salary_data.csv")
```

```python
#Comparison with Linear Regression wiht Scikit-learn
X=df.iloc[:,:-1].values
Y=df.iloc[:,1].values

x_train, x_test, y_train, y_test = tts(X,Y,test_size=0.33, random_state=2)
model=LinearRegression()
model.fit(x_train, y_train)
test_data_prediction = model.predict(x_test)

mae = mean_absolute_error(y_test, test_data_prediction)
mse= mean_squared_error(y_test, test_data_prediction)
r2 = r2_score(y_test, test_data_prediction)
print("MAE: ", mae)
print("MSE: ", mse)
print("R2: ", r2)
```

```python
#Comparison with Random Forest
from sklearn.ensemble import RandomForestRegressor
X=df.drop("Salary", axis=1)
Y=df["Salary"]
x_train, x_test, y_train, y_test = tts(X,Y, test_size=0.25, random_state=45)
rf_regressor = RandomForestRegressor(
    n_estimators = 500,
    max_depth = 10,
    min_samples_split = 5,
    min_samples_leaf = 2,
    random_state = 42
)
rf_regressor.fit(x_train, y_train)
y_pred = rf_regressor.predict(x_test)
print("MSE: ", mean_squared_error(y_test, y_pred))
print("R2 Score: ", r2_score(y_test, y_pred))
```

```python
#Comparison with XGBoost 
import xgboost
from xgboost import XGBRegressor
X = df.drop("Salary", axis=1)
Y = df["Salary"]

x_train, x_test, y_train, y_test = tts(X, Y, test_size = 0.25, random_state=45)

xgb = XGBRegressor(
    n_estimators = 300,
    learning_rate = 0.05,
    max_depth = 5,
    min_child_weight = 3,
    subsample = 0.8,
    colsample_bytree = 0.8,
)

xgb.fit(x_train, y_train)
y_pred = xgb.predict(x_test)

print("MSE: ", mean_squared_error(y_test, y_pred))
print("MAE: ", mean_absolute_error(y_test, y_pred))
print("R2_Score: ", r2_score(y_test, y_pred))
```
